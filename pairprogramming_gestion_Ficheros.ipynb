2. Abrir el archivo, creando un Dataframe. Aseguraros de que no incluye la columna "Unnamed: 0".

```python
df = pd.read_csv("videogames.csv",index_col = 0)
```

7. Guardar el Dataframe en formato Excel y después crear un Dataframe con el Excel. ¿EL Dataframe es similar al que teníais antes?

```python

archivo_excel = 'videogames.xlsx'
df.to_excel(archivo_excel, index=False)
df_nuevo = pd.read_excel(archivo_excel)

iguales = df.equals(df_nuevo)

iguales
```

9. ¿Tiene nulos o duplicados?

```python
nulos = df.isnull().sum()
print("Numero de registros nulos ",nulos)
duplicados = df.duplicated().sum()
print("Numero de registros duplicados ",duplicados)
```
